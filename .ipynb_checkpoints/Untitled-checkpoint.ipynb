```python
#Importando as bibliotecas
import requests
from bs4 import BeautifulSoup
```

```python
#escolhendo a pagina web que vamos fazer o scraping e colocando-a em uma variavel url

url ='https://www.asus.com/br/deals/mobile-handhelds/'
```

```python
#definindo o header

headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36"}
```

```python
# verificando se o site responde ao nosso Headers
response = requests.get(url, headers=headers)

# Verificar se a requisição foi bem-sucedida (código de status 200)
if response.status_code == 200:
    print("Requisição bem-sucedida!")
    # Exibir o conteúdo da resposta
    print(response.content)
else:
    print("Falha na requisição. Código de status:", response.status_code)
```

```python
site = requests.get(url, headers=headers)
soup = BeautifulSoup(site.content, 'html.parser')
celulares = soup.find_all('div', class_='ProductCardNormalStore2__viewBox__3NgQ2')
for item in celulares:
    
    item = item.find('a', class_='headingRow ProductCardNormalStore2__headingRow__3sb9D')
    
    if item:
        print(item.text)
#---------------------------------------------------------------------------
for config in celulares:
    
    config = config.find('div', class_='itemModelSpec ProductCardNormalStore2__itemModelSpec__cg96R')
    
    if config:
        print(config.text)
#---------------------------------------------------------------------------
for preco in celulares:
    
    preco = preco.find('div', class_='itemModelSpec ProductCardNormalStore2__itemModelSpec__cg96R')
    
    if preco:
        print(preco.text)


```
